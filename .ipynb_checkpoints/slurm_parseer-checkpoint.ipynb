```python
import matplotlib.pyplot as plt
import numpy as np
import math

import pickle5 as pickle

fig, plots = plt.subplots(1, 2, figsize=(10,4))

plots[0].grid(axis='y', alpha=0.75)
plots[0].set_xlabel('Iteration')
plots[0].set_ylabel('Error, a.u.')


plots[1].grid(axis='y', alpha=0.75)
plots[1].set_xlabel('Iteration')
plots[1].set_ylabel('Error, a.u.')



file_to_open= ["slurm-152738.out","slurm-152739.out"]
    
iteration=[]
errors=[]

file = open(file_to_open[0],'rb').readlines()
for line_number in range(len(file)):
    line=file[line_number].decode("utf-8")
    if len(line)>45 and line[:45]=="                         ADAPT-VQE iteration:":
        if float(file[line_number+4].decode("utf-8")[7:])>0:
            iteration.append(int(line[47:]))
            error_line=file[line_number+4].decode("utf-8")
            errors.append(float(error_line[7:]))



plots[0].plot(iteration,errors,linestyle='-', marker='.', color='g')
plots[0].grid( linestyle = '-', linewidth = 0.5)
plots[0].set_yscale("log")

iteration=[]
errors=[]

file = open(file_to_open[1],'rb').readlines()
for line_number in range(len(file)):
    line=file[line_number].decode("utf-8")
    if len(line)>45 and line[:45]=="                         ADAPT-VQE iteration:":
        if float(file[line_number+4].decode("utf-8")[7:])>0:
            iteration.append(int(line[47:]))
            error_line=file[line_number+4].decode("utf-8")
            errors.append(float(error_line[7:]))



plots[1].plot(iteration,errors,linestyle='-', marker='.', color='b')
plots[1].grid( linestyle = '-', linewidth = 0.5)
plots[1].set_yscale("log")

fig.savefig('random_Hamiltonians_error_vs_iteration')  

#plt.show()



# Set a clean upper y-axis limit.
```
